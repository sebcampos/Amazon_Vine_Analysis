```python
import pandas
```

```python
vines_df = pandas.read_csv("vine.csv")
```

```python

vines_df = vines_df.dropna()
vines_df = vines_df.drop_duplicates()
vines_df.head()
```

```python
votes_greater_than_20 = vines_df[vines_df.total_votes >= 20]
votes_greater_than_20
```

```python
greater_than_50 = votes_greater_than_20[votes_greater_than_20.helpful_votes / votes_greater_than_20.total_votes >= .5]
greater_than_50
```

```python
vine_paid = greater_than_50[greater_than_50.vine == "Y"]
vine_paid
```

```python
vine_not_paid = greater_than_50[greater_than_50.vine == "N"]
vine_not_paid
```

```python
total_number_of_reviews = vines_df.review_id.count()
total_number_of_reviews
```

```python
five_star = vines_df[vines_df.star_rating == 5][["star_rating"]].count()
five_star
```

```python
percentage_5_star_paid = vine_paid[vine_paid.star_rating == 5][["star_rating"]].count() / vine_paid.review_id.count()
print("total paid 5 star",vine_paid[vine_paid.star_rating == 5][["star_rating"]].count())
percentage_5_star_paid 
```

```python
percentage_5_star_not_paid = vine_not_paid[vine_not_paid.star_rating == 5][["star_rating"]].count() / vine_not_paid.review_id.count()
print("total unpaid 5 star",vine_not_paid[vine_not_paid.star_rating == 5][["star_rating"]].count())
percentage_5_star_not_paid
```

```python
summary_df = pandas.DataFrame({
    "Total (paid and unpaid)": str((five_star / total_number_of_reviews).item()).split(".")[2],
    "Percentage Vine Paid": str(percentage_5_star_paid.item()),
    "Percentage Vine Unpaid": str(percentage_5_star_not_paid.item())
}, index=["5 Star Percentage"])

summary_df
```
